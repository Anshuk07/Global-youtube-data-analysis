```python
#merge the created year/month/day columns and set the type to datetime
data['created_year'] = data['created_year'].astype(str).str.split('.').str[0] #to only get the year bc the type is float originally exp. 2006.0
data['created_month'] = data['created_month'].astype(str)
data['created_date'] = data['created_date'].astype(str).str.split('.').str[0]
```

```python
data['creation_date'] = data['created_year'] + '/' + data['created_month']  + '/' + data['created_date']
```

```python
data = data.drop(columns=['created_year', 'created_month', 'created_date'], axis=1)
```

```python
data['creation_date'].isnull().sum()
```

```python
data['creation_date'] = pd.to_datetime(data['creation_date'], errors='coerce')
```

```python
data.isnull().sum()
data.rename(columns={'video views':'video_views'}, inplace=True)
```

```python
#rename all the col name to lower case
data.columns = [s.lower() for s in data.columns.tolist()]
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression

#create a object
label_encoder = LabelEncoder()
```

```python
train_data = data.dropna(subset=['category'])
predict_data = data[data['category'].isnull()]
```

```python
features = ['encoded_youtuber', 'encoded_title']
target = 'category'
```

```python
logreg = LogisticRegression()
```

```python
logreg.fit(train_data[features], train_data[target])
```

```python
predict_categories = logreg.predict(predict_data[features])
```

```python
data.loc[data['category'].isnull(), 'category'] = predict_categories
```
